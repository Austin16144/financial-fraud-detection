# Week 1: Exploratory Data Analysis (EDA)

**Project:** Advanced Financial Fraud Detection

**Goal:** Understand the credit card dataset and find patterns useful for fraud detection modeling.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Settings
sns.set(style='whitegrid')
```

```python
# Load dataset
df = pd.read_csv('../data/creditcard.csv')

# Basic info
df.shape, df.info()
```

```python
print(df.columns.tolist())
df.columns = df.columns.str.strip()  # removes extra spaces
```

```python
# Class distribution
print(df['Class'].value_counts())
print(df['Class'].value_counts(normalize=True) * 100)

# Plot class distribution
sns.countplot(x='Class', data=df)
plt.title('Class Distribution: Fraud vs Non-Fraud')
plt.show()
```

```python
# Null values
print(df.isnull().sum())

# Duplicate rows
print('Duplicate Rows:', df.duplicated().sum())
```

```python
# Transaction Amount
sns.histplot(df['Amount'], bins=100, kde=True)
plt.title('Transaction Amount Distribution')
plt.show()

# Transaction Time
sns.histplot(df['Time'], bins=100, kde=True)
plt.title('Transaction Time Distribution')
plt.show()
```

```python
# Correlation Heatmap
plt.figure(figsize=(14, 10))
sns.heatmap(df.corr(), cmap='coolwarm', vmax=0.8, square=True)
plt.title('Correlation Heatmap')
plt.show()
```

```python
# Boxplots for selected features
features = ['V1', 'V2', 'V3', 'V4', 'V14', 'V17', 'Amount']
for col in features:
    sns.boxplot(x='Class', y=col, data=df)
    plt.title(f'{col} vs Class')
    plt.show()
```

##  Summary
- Dataset has **284,807** transactions, with ~0.17% fraud cases.
- No missing values found.
- Features like **V14**, **V17**, and **Amount** show potential separation between classes.
- Clear class imbalance needs handling in Week 2.
